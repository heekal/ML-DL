### Approaching a Machine Learning Problem

Jangan terburu-buru menggunakan algoritma. *Machine learning* hanyalah sebagian kecil dari proses penyelesaian masalah yang lebih besar. Berikut adalah langkah krusial yang harus dilakukan:

* **Tentukan Tujuan dan Metrik:** Pastikan Anda tahu masalah spesifik yang ingin diselesaikan, cara mengukur kesuksesannya, dan ketersediaan data untuk evaluasi.
* **Uji Kelayakan Bisnis:** Evaluasi potensi dampaknya. Jika sebuah "model sempurna" hanya memberikan keuntungan finansial yang kecil, proyek tersebut mungkin tidak layak dikerjakan. Fokuslah pada masalah yang memberikan dampak signifikan.
* **Gunakan Siklus Umpan Balik:** Pembangunan model adalah proses berulang yang melibatkan pengumpulan, pembersihan, pemodelan, dan analisis. Evaluasi kesalahan model untuk mengetahui data apa yang kurang.

> **Kesimpulan Penting:** Menambah data yang relevan atau merumuskan ulang masalah sering kali memberikan hasil yang jauh lebih baik daripada menghabiskan waktu tanpa henti untuk mengoptimalkan parameter algoritma.

### Humans in the Loop

Anda perlu mempertimbangkan apakah sistem memerlukan campur tangan manusia. Tidak semua proses harus sepenuhnya otomatis.

1. Otomatisasi Kasus Sederhana: Algoritma dapat difokuskan untuk menangani sebagian besar kasus yang mudah secara otomatis. Mengotomatiskan 10% hingga 90% pekerjaan dasar ini sudah cukup untuk menghemat waktu dan biaya secara signifikan.

2. Eskalasi Kasus Rumit: Untuk keputusan yang tidak pasti atau membutuhkan tingkat akurasi yang sangat tinggi (seperti diagnosis medis), algoritma dapat menyerahkan keputusan akhir tersebut kepada manusia.

### From Prototype to Production

Python dan *scikit-learn* sangat ideal untuk eksperimen dan pembuatan prototipe cepat. Namun, saat mengimplementasikan model ke sistem produksi, terdapat standar dan tantangan yang berbeda:

* **Pemisahan Bahasa Pemrograman:** Sangat umum bagi tim data untuk menguji ide menggunakan Python/R, lalu tim produksi menulis ulang model akhir tersebut menggunakan bahasa berkinerja tinggi (seperti Go, C++, Java, atau Scala) agar lebih tangguh dan mudah diskalakan.
* **Prioritas Rekayasa Perangkat Lunak:** Berbeda dengan skrip analisis biasa, sistem produksi menuntut keandalan yang tinggi, kecepatan proses (*runtime*), efisiensi memori, dan prediktabilitas.
* **Utamakan Kesederhanaan:** Evaluasi secara kritis setiap tahap pemrosesan data Anda. Pastikan manfaat dari setiap komponen sepadan dengan kerumitannya agar sistem tetap mudah dipelihara dan Anda terhindar dari tumpukan beban teknis (*technical debt*).

### Testing Production Systems

Menguji model menggunakan data historis yang sudah dikumpulkan (*offline evaluation*) hanyalah langkah awal. Untuk sistem yang berinteraksi langsung dengan pengguna, pengujian langsung di dunia nyata (*online testing*) sangatlah penting:

* **Pentingnya Pengujian Langsung:** Mengubah algoritma (seperti sistem rekomendasi atau pencarian) dapat mengubah perilaku pengguna secara tak terduga. Evaluasi *online* membantu mengukur dampak nyata ini.
* ***A/B Testing*:** Ini adalah metode pengujian *online* yang paling umum. Pengguna dibagi secara acak tanpa sepengetahuan mereka; sebagian menggunakan algoritma baru (A), dan sisanya menggunakan sistem lama (B). Metrik keberhasilan keduanya kemudian dicatat dan dibandingkan.
* **Metode Lanjutan:** Terdapat juga metode pengujian *online* yang lebih canggih dari *A/B testing* untuk mengoptimalkan sistem, seperti *bandit algorithms*.

#### Scaling to Larger Datasets

Jika ukuran data sangat besar hingga tidak muat di dalam memori (RAM), Anda tidak bisa lagi menyimpannya dalam format NumPy *array* atau SciPy *sparse matrix* biasa. Untuk mengatasi batasan ini, terdapat dua strategi utama:

* ***Out-of-Core Learning* (Satu Komputer):** Metode ini melatih model pada satu komputer dengan cara membaca data secara bertahap (per sampel atau per bongkahan/*chunk*) langsung dari *hard disk* atau jaringan. Setelah model diperbarui menggunakan *chunk* tersebut, data di RAM akan dihapus untuk digantikan dengan *chunk* berikutnya. Beberapa model di *scikit-learn* sudah mendukung fitur ini, namun kekurangannya adalah waktu proses yang lama dan tidak semua algoritma bisa diterapkan dengan cara ini.
* **Paralelisasi Berbasis *Cluster* (Banyak Komputer):** Strategi ini mendistribusikan data ke beberapa mesin sekaligus dalam sebuah *cluster* komputer untuk diproses secara bersamaan. Pendekatan ini jauh lebih cepat dan kapasitas datanya hanya dibatasi oleh ukuran *cluster* itu sendiri. Salah satu platform terpopuler untuk kebutuhan ini adalah Apache Spark (menggunakan paket MLLib) yang berjalan di atas Hadoop, atau bisa juga menggunakan Vowpal Wabbit (`vw`) jika Anda tidak ingin membangun infrastruktur *cluster* yang terlalu rumit.
