### Import Library

```python
!pip install -q umap-learn

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
import umap
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import warnings
warnings.filterwarnings('ignore')

print(f"TF version: {tf.__version__}")
print(f"GPU available: {tf.config.list_physical_devices('GPU')}")
```

### Preprocessing: Handle Missing Values & Drop ID

CREDIT_LIMIT: 1 missing → isi median </br>
MINIMUM_PAYMENTS: 313 missing → isi median (distribusi skewed) </br>
CUST_ID bukan fitur, drop.

```python
df_clean = df.copy()
df_clean.drop(columns=['CUST_ID'], inplace=True)

df_clean['CREDIT_LIMIT'].fillna(df_clean['CREDIT_LIMIT'].median(), inplace=True)
df_clean['MINIMUM_PAYMENTS'].fillna(df_clean['MINIMUM_PAYMENTS'].median(), inplace=True)

print("Missing after imputation:", df_clean.isnull().sum().sum())
print("Shape:", df_clean.shape)
```

### Outlier Handling (IQR Capping)

Pakai capping (winsorizing) bukan drop — preserves data. </br>
Clip ke [Q1 - 1.5*IQR, Q3 + 1.5*IQR] per kolom numerik. </br>
Lebih aman daripada drop untuk dataset 8950 baris.

```python
df_capped = df_clean.copy()
for col in df_capped.columns:
  Q1 = df_capped[col].quantile(0.25)
  Q3 = df_capped[col].quantile(0.75)
  IQR = Q3 - Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR
  df_capped[col] = df_capped[col].clip(lower, upper)

print("Outlier capping done. Shape:", df_capped.shape)
```

### Feature Engineering

Tambah fitur turunan yang relevan untuk spending/payment behavior. </br>
Ratio lebih informatif daripada nilai absolut untuk clustering. </br>

```python
df_feat = df_capped.copy()

# Rasio penggunaan kartu
df_feat['BALANCE_TO_LIMIT']       = df_feat['BALANCE'] / (df_feat['CREDIT_LIMIT'] + 1)
df_feat['PURCHASES_TO_LIMIT']     = df_feat['PURCHASES'] / (df_feat['CREDIT_LIMIT'] + 1)
df_feat['CASH_TO_PURCHASES']      = df_feat['CASH_ADVANCE'] / (df_feat['PURCHASES'] + 1)
df_feat['PAYMENT_TO_BALANCE']     = df_feat['PAYMENTS'] / (df_feat['BALANCE'] + 1)
df_feat['INSTALLMENT_RATIO']      = df_feat['INSTALLMENTS_PURCHASES'] / (df_feat['PURCHASES'] + 1)

# Clip rasio agar tidak unbounded
for col in ['BALANCE_TO_LIMIT', 'PURCHASES_TO_LIMIT', 'CASH_TO_PURCHASES',
            'PAYMENT_TO_BALANCE', 'INSTALLMENT_RATIO']:
  df_feat[col] = df_feat[col].clip(0, 5)

print("Features after engineering:", df_feat.shape[1])
print(df_feat.head(2))
```

### Normalization

RobustScaler lebih tahan outlier residual vs StandardScaler. </br>
Setelah ini, semua fitur dalam skala yang sama. </br>

```python
scaler = RobustScaler()
X_scaled = scaler.fit_transform(df_feat)

print("Scaled shape:", X_scaled.shape)
print("Sample mean (should be ~0):", X_scaled.mean(axis=0)[:5].round(3))
```

### EDA Visual: Correlation Heatmap

Lihat multikolinearitas sebelum masuk model. </br>
Beberapa fitur sangat correlated (purchases variants) </br>
autoencoder yang akan handle dimensionality reduction-nya.

```python
plt.figure(figsize=(14, 10))
corr = df_feat.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=False, cmap='coolwarm',
            linewidths=0.3, vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix', fontsize=14)
plt.tight_layout()
plt.show()
```

### Deep Learning: Autoencoder Architecture

Autoencoder = encoder compresses input ke latent space, </br>
decoder reconstruct kembali. Latent space = representasi </br>
learned yang dipakai untuk clustering. </br>
</br>
Dimensi input: 22 fitur → bottleneck 8 (latent dim). </br>
Pakai Batch Normalization + Dropout untuk regularization.</br>
Colab free: model kecil, cukup GPU/CPU.</br>

```python
INPUT_DIM = X_scaled.shape[1]  # 22
LATENT_DIM = 8
HIDDEN_DIMS = [128, 64, 32]

def build_autoencoder(input_dim, latent_dim, hidden_dims):
  # --- Encoder ---
  inputs = keras.Input(shape=(input_dim,))
  x = inputs
  for dim in hidden_dims:
    x = layers.Dense(dim)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Dropout(0.2)(x)
  latent = layers.Dense(latent_dim, name='latent')(x)

  # --- Decoder ---
  y = latent
  for dim in reversed(hidden_dims):
    y = layers.Dense(dim)(y)
    y = layers.BatchNormalization()(y)
    y = layers.Activation('relu')(y)
    y = layers.Dropout(0.2)(y)
  outputs = layers.Dense(input_dim, activation='linear', name='reconstruction')(y)

  autoencoder = keras.Model(inputs, outputs, name='autoencoder')
  encoder     = keras.Model(inputs, latent, name='encoder')
  return autoencoder, encoder

autoencoder, encoder = build_autoencoder(INPUT_DIM, LATENT_DIM, HIDDEN_DIMS)
autoencoder.summary()
```

### Training Autoencoder

Loss: MSE antara input dan rekonstruksi. </br>
ReduceLROnPlateau: turunkan LR kalau val_loss stagnan. </br>
EarlyStopping: stop kalau tidak improve 10 epoch. </br>
batch_size=256 hemat memory vs 512 tidak pengaruh banyak. </br>

```python
autoencoder.compile(
  optimizer=keras.optimizers.Adam(learning_rate=1e-3),
  loss='mse'
)

callbacks = [
  keras.callbacks.EarlyStopping(monitor='val_loss', patience=10,
                                 restore_best_weights=True, verbose=1),
  keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                                     patience=5, verbose=1, min_lr=1e-6)
]

history = autoencoder.fit(
  X_scaled, X_scaled,
  epochs=100,
  batch_size=256,
  validation_split=0.1,
  callbacks=callbacks,
  verbose=1
)
```

### Plot Training Loss

Cek apakah model converge dan tidak overfit. </br>
Val loss harus mengikuti train loss, gap kecil = bagus. </br>

```python
plt.figure(figsize=(8, 4))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('Autoencoder Training Loss')
plt.legend()
plt.tight_layout()
plt.show()

# Reconstruction quality check
recon = autoencoder.predict(X_scaled, batch_size=512, verbose=0)
recon_error = np.mean((X_scaled - recon) ** 2, axis=1)
print(f"Mean Reconstruction Error: {recon_error.mean():.4f} ± {recon_error.std():.4f}")
```

### Extract Latent Representations

Pakai encoder (bukan full autoencoder) untuk dapatkan </br>
latent space 8-dimensi. Ini yang akan di-cluster. </br>

```python
X_latent = encoder.predict(X_scaled, batch_size=512, verbose=0)
print("Latent space shape:", X_latent.shape)
```

### Elbow Method + Silhouette untuk Pilih K

Elbow: titik dimana WCSS mulai tidak turun signifikan. </br>
Silhouette score: makin tinggi makin baik (range -1 to 1). </br>
Range K=2..9 cukup; lebih dari 9 cluster sulit diinterpretasi. </br>

```python
inertia = []
silhouette_scores = []
db_scores = []
K_range = range(2, 10)

for k in K_range:
  km = KMeans(n_clusters=k, random_state=42, n_init=10)
  labels = km.fit_predict(X_latent)
  inertia.append(km.inertia_)
  silhouette_scores.append(silhouette_score(X_latent, labels, sample_size=3000))
  db_scores.append(davies_bouldin_score(X_latent, labels))
  print(f"K={k} | Inertia={km.inertia_:.1f} | Silhouette={silhouette_scores[-1]:.4f} | DB={db_scores[-1]:.4f}")

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].plot(K_range, inertia, 'bo-')
axes[0].set_title('Elbow Method (WCSS)')
axes[0].set_xlabel('K')
axes[0].set_ylabel('Inertia')

axes[1].plot(K_range, silhouette_scores, 'go-')
axes[1].set_title('Silhouette Score (higher=better)')
axes[1].set_xlabel('K')
axes[1].set_ylabel('Score')

axes[2].plot(K_range, db_scores, 'ro-')
axes[2].set_title('Davies-Bouldin (lower=better)')
axes[2].set_xlabel('K')
axes[2].set_ylabel('Score')

plt.tight_layout()
plt.show()
```

### Final Clustering dengan K optimal

```python
OPTIMAL_K = 3

kmeans_final = KMeans(n_clusters=OPTIMAL_K, random_state=42, n_init=20)
cluster_labels = kmeans_final.fit_predict(X_latent)

df_result = df.copy()
df_result['CLUSTER'] = cluster_labels

print("Cluster distribution:")
print(df_result['CLUSTER'].value_counts().sort_index())
print(f"\nFinal Silhouette Score: {silhouette_score(X_latent, cluster_labels):.4f}")
print(f"Final Davies-Bouldin:   {davies_bouldin_score(X_latent, cluster_labels):.4f}")
print(f"Final Calinski-Harabasz: {calinski_harabasz_score(X_latent, cluster_labels):.4f}")
```

### UMAP Visualization (2D)

PCA cepat tapi linear. UMAP capture non-linear structure </br>
lebih informatif untuk DL latent space. </br>
n_neighbors=15 & min_dist=0.1 adalah default yang baik. </br>

```python
reducer = umap.UMAP(n_components=2, n_neighbors=15, min_dist=0.1,
                    random_state=42, verbose=False)
X_umap = reducer.fit_transform(X_latent)

plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_umap[:, 0], X_umap[:, 1],
                      c=cluster_labels, cmap='tab10', s=8, alpha=0.7)
plt.colorbar(scatter, label='Cluster')
plt.title(f'UMAP Projection of Latent Space (K={OPTIMAL_K})')
plt.xlabel('UMAP-1')
plt.ylabel('UMAP-2')
plt.tight_layout()
plt.show()
```

### PCA Visualization (2D) sebagai pembanding

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_latent)
print(f"Explained variance ratio: {pca.explained_variance_ratio_.sum():.2%}")

plt.figure(figsize=(10, 7))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1],
                      c=cluster_labels, cmap='tab10', s=8, alpha=0.7)
plt.colorbar(scatter, label='Cluster')
plt.title(f'PCA Projection of Latent Space (K={OPTIMAL_K})')
plt.xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%})')
plt.ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%})')
plt.tight_layout()
plt.show()
```

### Cluster Profiling: Mean per Cluster

supaya interpretasi masuk akal (nilai aktual, bukan scaled). </br>

```python
key_features = [
  'BALANCE', 'PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT',
  'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT',
  'PURCHASES_FREQUENCY', 'CASH_ADVANCE_FREQUENCY',
  'BALANCE_FREQUENCY', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES'
]

profile = df_result.groupby('CLUSTER')[key_features].mean().round(2)
profile['COUNT'] = df_result['CLUSTER'].value_counts().sort_index()
print(profile.T)
```

### Radar Chart per Cluster

Visual intuitif untuk compare antar cluster. </br>
Normalize ke 0-1 supaya skala bisa dibandingkan. </br>

```python
from matplotlib.patches import FancyArrowPatch
import matplotlib.patches as mpatches

radar_features = [
  'BALANCE', 'PURCHASES', 'CASH_ADVANCE',
  'CREDIT_LIMIT', 'PAYMENTS', 'PRC_FULL_PAYMENT',
  'PURCHASES_FREQUENCY', 'CASH_ADVANCE_FREQUENCY'
]

# Normalize 0-1 dari data asli
profile_norm = (profile[radar_features] - profile[radar_features].min()) / \
               (profile[radar_features].max() - profile[radar_features].min() + 1e-9)

angles = np.linspace(0, 2 * np.pi, len(radar_features), endpoint=False).tolist()
angles += angles[:1]  # close polygon

fig, axes = plt.subplots(1, OPTIMAL_K, figsize=(5 * OPTIMAL_K, 5),
                          subplot_kw=dict(polar=True))
colors = plt.cm.tab10(np.linspace(0, 0.9, OPTIMAL_K))

for i, ax in enumerate(axes):
  values = profile_norm.iloc[i].tolist()
  values += values[:1]
  ax.plot(angles, values, color=colors[i], linewidth=2)
  ax.fill(angles, values, color=colors[i], alpha=0.25)
  ax.set_xticks(angles[:-1])
  ax.set_xticklabels(radar_features, size=7)
  ax.set_title(f'Cluster {i}', size=12, pad=15)
  ax.set_ylim(0, 1)

plt.suptitle('Cluster Radar Profiles', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```
